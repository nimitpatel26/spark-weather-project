```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sf_df = pd.read_csv("s3://spark-weather-project/data/sf_test.csv")
seattle_df = pd.read_csv("s3://spark-weather-project/data/seattle_test.csv")
```

```python

def plot(df, column_name_actual, column_name_forecast):
    x = df["Row Number"]
    y1 = df[column_name_actual]
    y2 = df[column_name_forecast]

    s = [8 for i in range(len(y1))]
    plt.plot(x, y1, color = "orange", label = "Actual")
    plt.scatter(x, y2, s = s, label = "Predicted")
    plt.xlim([350000, 350300])
    plt.ylim([40, 70])

    plt.title("Actual vs. Predicted Temperature (" + column_name_forecast + ")")
    plt.xlabel('Day Number')
    plt.ylabel('Temperature')
    plt.legend()
    plt.show()
```

```python
print("=========================================")
print("San Francisco")
print("=========================================")
plot(sf_df, "actual_24hour", "forecast_24hour")
plot(sf_df, "actual_72hour", "forecast_72hour")
```
